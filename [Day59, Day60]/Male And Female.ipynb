```python
import cv2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import os
import numpy as np
import tensorflow as tf
import tensorflowjs as tfjs
```

```python
DATADIR = "/Users/Th3D4rKnight/Desktop/GUFD/3cameras_cropped/C1"
CATEGORIES = ['male','female']
```

```python
training_data=[]
def prepare_training_data():
    for category in CATEGORIES:
        path = os.path.join(DATADIR,category)
        class_num = CATEGORIES.index(category)
        for img in os.listdir(path):
                img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                IMG_SIZE = 50
                new_image = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_image,class_num])    
```

```python
prepare_training_data()
```

```python
print(len(training_data))
```

```python
import random
```

```python
random.shuffle(training_data)
```

```python
X = [] 
y = []
```

```python
for features,labels in training_data:
    X.append(features)
    y.append(labels)
```

```python
X = np.array(X).reshape(-1,50,50,1)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout,Activation,Flatten,Conv2D,MaxPooling2D

import pickle

X = X/255.0
model = Sequential()
model.add(Conv2D(64,(3,3),input_shape = X.shape[1:]))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(64,(3,3)))
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(64))

model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss="binary_crossentropy",optimizer="adam",metrics=['accuracy'])
model.fit(X,y,batch_size=32,epochs=150,validation_split=0.1)
```
