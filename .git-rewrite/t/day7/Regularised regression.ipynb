# Regularised Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
% matplotlib inline
import pandas as pd
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
n_l = 100;
np.random.seed(42)
rng  = np.random.rand(n_l)*10;
y = 0.5 * rng + 2 * np.random.rand(n_l)

lr = LinearRegression()
lr.fit(rng.reshape(-1,1),y);
lr_pred  = lr.predict(rng.reshape(-1,1))

plt.figure(figsize=(12,10))
plt.scatter(rng,y)
plt.show()
print(lr.coef_);
```

```python
idx = rng.argmin()
y[idx] = 200
```

```python
o_lr = LinearRegression(normalize=True)
o_lr.fit(rng.reshape(-1,1),y)
o_model_pred = o_lr.predict(rng.reshape(-1,1))

plt.figure(figsize=(12,10))
plt.scatter(rng,y);
plt.plot(rng,o_model_pred);
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso_mod = Lasso(alpha=0.5, normalize=True)
lasso_mod.fit(rng.reshape(-1,1),y)
lasso_model_pred = lasso_mod.predict(rng.reshape(-1,1))

plt.figure(figsize=(12,10))
plt.scatter(rng,y)
plt.plot(rng,lasso_model_pred);
print(lasso_mod.coef_)
```
